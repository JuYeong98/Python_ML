```python

# -*- coding: utf-8 -*-

#  
import pandas as pd
import matplotlib.pyplot as plt

# matplotlib     
from matplotlib import font_manager, rc
font_path = 'C:/Users/chan/5674-833/part4/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

plt.style.use('ggplot')   #   
plt.rcParams['axes.unicode_minus']=False   #    

# Excel    
df = pd.read_excel(r'C:/Users/chan/5674-833/part4/.xlsx',convert_float= True)#  
df = df.loc[5:9]
df.drop(' (h)', axis='columns', inplace=True)
df.set_index(' ', inplace=True)
df = df.T 

# () 
df = df.rename(columns={'':''})
df[' - 1'] = df[''].shift(1)
df[''] = ((df[''] / df[' - 1']) - 1) * 100      

# 2  
ax1 = df[['','']].plot(kind='bar', figsize=(20, 10), width=0.7, stacked=True)  
ax2 = ax1.twinx()
ax2.plot(df.index, df., ls='--', marker='o', markersize=20, 
         color='green', label=' (%)')  

ax1.set_ylim(0, 500)
ax2.set_ylim(-50, 50)

ax1.set_xlabel('', size=20)
ax1.set_ylabel('( KWh)')
ax2.set_ylabel('  (%)')

plt.title('   (1990 ~ 2016)', size=30)
ax1.legend(loc='upper left')

plt.show()
```
